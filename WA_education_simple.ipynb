# Model Making!

```python
# ENGL Model
engl_df = education_df[education_df["TestSubject"]=="ELA"]
engl_df = engl_df.drop("TestSubject", axis=1)

# Math Model
math_df = education_df[education_df["TestSubject"]=="Math"]
math_df = math_df.drop("TestSubject", axis=1)
```

```python
# Engl Model input
engl_all_x = engl_df.drop("PercentMetStandardSuppressed", axis=1).to_numpy()

# Math Model input
math_all_x = math_df.drop("PercentMetStandardSuppressed", axis=1).to_numpy()
```

```python
from sklearn.preprocessing import RobustScaler

# Data Scaling
eng_scaling_df = engl_df
math_scaling_df = math_df

eng_transformer = RobustScaler().fit(eng_scaling_df)
math_transformer = RobustScaler().fit(math_scaling_df)

eng_transformed = eng_transformer.transform(eng_scaling_df)
math_transformed = math_transformer.transform(math_scaling_df)

eng_data_scaled_df = pd.DataFrame(eng_transformed, columns=engl_df.columns)
math_data_scaled_df = pd.DataFrame(math_transformed, columns=math_df.columns)

# scale to z-scores

# engl model output
engl_all_y = eng_data_scaled_df["PercentMetStandardSuppressed"]

# math model output
math_all_y = math_data_scaled_df["PercentMetStandardSuppressed"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
# Engl Model splitting
x_trval_E, x_te_E, y_trval_E, y_te_E = train_test_split(engl_all_x, engl_all_y, test_size=0.2, random_state=0)
x_tr_E, x_val_E, y_tr_E, y_val_E = train_test_split(x_trval_E, y_trval_E, test_size=0.2, random_state=0)

# Math Model Splitting
x_trval_M, x_te_M, y_trval_M, y_te_M = train_test_split(math_all_x, math_all_y, test_size=0.2, random_state=0)
x_tr_M, x_val_M, y_tr_M, y_val_M = train_test_split(x_trval_M, y_trval_M, test_size=0.2, random_state=0)
```

```python
from sklearn import linear_model
h = linear_model.LinearRegression()
```

```python
# Engl: Train (fit) the model
h.fit(x_tr_E, y_tr_E)

# Math: Train (fit) the model
h.fit(x_tr_M, y_tr_M)
```

```python
from sklearn.metrics import mean_squared_error

# Engl Predictions
pred_val_E = h.predict(x_val_E)
engl_predicted = pd.DataFrame(pred_val_E)

# Math Predictions
pred_val_M = h.predict(x_val_M)
math_predicted = pd.DataFrame(pred_val_M)

print(h.coef)
```

```python
# Engl "Scores" with validation set:
print("ENGL Prediction Accuracy:")
print("Val R^2 = ", h.score(x_val_E, y_val_E))
print("Val RMSE = ", mean_squared_error(y_val_E, pred_val_E, squared=False))

# Math "Scores" with validation set:
print("\nMath Prediction Accuracy:")
print("Val R^2 = ", h.score(x_val_M, y_val_M))
print("Val RMSE = ", mean_squared_error(y_val_M, pred_val_M, squared=False))
```
